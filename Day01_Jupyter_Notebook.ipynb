```python
#Let's create some new features with help of existing feature:
features =[]
dtypes =[]
count =[]
uniques =[]
missing = []
missing_percentage=[]

for j in df.columns:
    features.append(j)
    count.append(len(df[j]))
    missing.append(df[j].isnull().sum())
    uniques.append(df[j].nunique())
    dtypes.append(df[j].dtypes)
    missing_percentage.append((df[j].isnull().sum()/df.shape[0])*100)
    
dataframe = pd.DataFrame({'feature':features,
                         'count':count,
                         'missing':missing,
                         'nunique':uniques,
                         'dtypes':dtypes,
                         'missing_percentage':missing_percentage})

dataframe.set_index('feature')
```
