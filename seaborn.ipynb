#jointplot - allows to study the relationship between two numeric variables. The centrail chart display their correlation. It is usually a scatterplot, a hexbinplot, a 2D hietogram or 2D density plot

### Box plot
A box or whisker plot is a graph that represents information from a five number summary

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df)#doing on its own
```

```python
sns.boxplot(data=df, orient ='v')
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow',hue='smoker')
```

### violin plot
It helps us to see both the distribution of data in terms of KDE and boxplot

```python
sns.violinplot(data=df)
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```

```python
df1 = df.select_dtypes([np.int,np.float])
for i, col in enumerate(df1.columns):
    plt.figure(i)
    plt.style.use('ggplot')
    sns.countplot(x=col, data=df1)
    plt.xticks(size=15,rotation=90)
    plt.tight_layout()
```

```python
for i in numeric_data:
    sns.displot(data=df,x=i,height=4,aspect=2,color='#158685')
```

```python
pvcap={'Germany':20000,'Japan':19000,"USA":18000,'china':17000}
countries = list(pvcap.keys())
cap = list(pvcap.values())
plt.bar(countries,cap,color='blue',width=0.8)
plt.title('Solar Usage')
plt.xlabel('countreis')
plt.ylabel('capacity')
```

```python
pvcap={'Germany':20000,'Japan':19000,"USA":18000,'china':17000}
countries = list(pvcap.keys())
cap = list(pvcap.values())
plt.bar(countries,cap,color='blue',width=0.8)
plt.title('Solar Usage')
plt.xlabel('countreis')
plt.ylabel('capacity')
for i, v in enumerate(cap):
    #i is the index v is the value
    plt.text(i, v+1000, str(v), ha='center',weight='bold',size=10)
```
