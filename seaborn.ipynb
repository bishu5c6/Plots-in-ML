```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
#correlation with heatnap
df.corr()
```

```python
sns.heatmap(df.corr())
```

#jointplot - allows to study the relationship between two numeric variables. The centrail chart display their correlation. It is usually a scatterplot, a hexbinplot, a 2D hietogram or 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#categorical Plots
sns.countplot(y='sex',data=df) #by default it takes as x-axis
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df,hue='sex')
```

```python
df.head(2)
```

```python
#in barplot we have to give both x and y value
sns.barplot(x='sex',y='tip',data=df)
```

```python
sns.barplot(x='sex',y='size',data=df)
```

### Box plot
A box or whisker plot is a graph that represents information from a five number summary

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df)#doing on its own
```

```python
sns.boxplot(data=df, orient ='v')
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow',hue='smoker')
```

### violin plot
It helps us to see both the distribution of data in terms of KDE and boxplot

```python
sns.violinplot(data=df)
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
