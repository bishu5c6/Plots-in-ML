```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
#correlation with heatnap
df.corr()
```

```python
sns.heatmap(df.corr())
```

#jointplot - allows to study the relationship between two numeric variables. The centrail chart display their correlation. It is usually a scatterplot, a hexbinplot, a 2D hietogram or 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```
